```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
vx_init = np.loadtxt('../vxI.txt')
vx_final = np.loadtxt('../vxF.txt')

# Plot for initial x-velocity distribution
plt.figure(1)
plt.hist(vx_init, bins=25, alpha=0.7, color='blue')
plt.xlabel('$v_x$ (m/s)')
plt.ylabel('Probability Density')
plt.title('Initial x-velocity distribution')

# Plot for final x-velocity distribution
plt.figure(2)
plt.hist(vx_final, bins=25, alpha=0.7, color='green')
plt.xlabel('$v_x$ (m/s)')
plt.ylabel('Probability Density')
plt.title('Final x-velocity distribution')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load initial and final velocity magnitudes
vmag_init = np.loadtxt('../vmagI.txt')
vmag_final = np.loadtxt('../vmagF.txt')

# Plot the initial speed distribution
plt.figure(1)
plt.hist(vmag_init, bins=25, alpha=0.7, color='blue')
plt.xlabel('Speed (m/s)')
plt.ylabel('Probability Density')
plt.title('Initial speed distribution (N=10000)')

# Plot the final speed distribution
plt.figure(2)
plt.hist(vmag_final, bins=25, alpha=0.7, color='green')
plt.xlabel('Speed (m/s)')
plt.ylabel('Probability Density')
plt.title('Final speed distribution (after 1000 steps)')

plt.show()
```
